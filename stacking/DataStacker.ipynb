```python
import pandas as pd
from dateutil.parser import parse
```

```python
df = pd.read_csv("../classification/Data/labeledTransactions.csv")
```

```python
# Calculate month based on date
def getDate(date):
    datetime = parse(date)
    return datetime.isocalendar()[1]

df['created_week'] = df['created_time'].apply(getDate)
df
```

```python
new_df = df.groupby(['created_week', 'category']).size().unstack(fill_value=0)
```

```python
new_df.to_csv("stackedTransactions.csv")
```
